```python
import cv2

# Load classifier (USE FULL PATH)
body_classifier = cv2.CascadeClassifier(
    '/Users/mayurimamdi/Downloads/haarcascade_fullbody.xml'
)

if body_classifier.empty():
    print("Error: Classifier not loaded")
    exit()

# Load video
cap = cv2.VideoCapture('/Users/mayurimamdi/Downloads/walking.avi')

if not cap.isOpened():
    print("Error: Video not loaded")
    exit()

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    bodies = body_classifier.detectMultiScale(gray, 1.2, 3)

    for (x, y, w, h) in bodies:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0,255,0), 2)

    cv2.imshow('Body Detection', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import time
import numpy as np

# Create our car classifier
car_classifier = cv2.CascadeClassifier('/Users/mayurimamdi/Downloads/haarcascade_car.xml')

# Initiate video capture
cap = cv2.VideoCapture('/Users/mayurimamdi/Downloads/cars.avi')

# ⏱️ Start timer
start_time = time.time()

# Loop once video is successfully loaded
while cap.isOpened():
    
    time.sleep(.05)

    # Read frame
    ret, frame = cap.read()

    if not ret:
        break

    # ⏱️ Stop after 5 seconds
    if time.time() - start_time > 5:
        print("Stopping after 5 seconds")
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
   
    # Detect cars
    cars = car_classifier.detectMultiScale(gray, 1.4, 2)
    
    # Draw rectangles
    for (x,y,w,h) in cars:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 255), 2)

    # Show frame (moved outside loop for better performance)
    cv2.imshow('Cars', frame)

    # Press Enter key to stop manually
    if cv2.waitKey(1) == 13:
        break

cap.release()
cv2.destroyAllWindows()
```
