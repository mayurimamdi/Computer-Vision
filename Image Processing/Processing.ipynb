```python
import numpy as np
from PIL import Image, ImageFilter, ImageEnhance
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

# ─────────────────────────────────────────────
# STEP 1: LOAD THE IMAGE
# ─────────────────────────────────────────────
# An image is just a 3D array of numbers:
#   - Height x Width x 3 (the 3 = Red, Green, Blue channels)
#   - Each number is 0–255 (0 = dark, 255 = bright)

img = Image.open("/Users/mayurimamdi/Downloads/Modi.jpg")       
img_array = np.array(img)                  

print("=" * 50)
print("📷 IMAGE LOADED!")
print(f"   Size      : {img.width} x {img.height} pixels")
print(f"   Array shape: {img_array.shape}  → (Height, Width, RGB)")
print(f"   Top-left pixel (R,G,B): {img_array[0, 0]}")
print(f"   This pixel is sky blue ☁️")
print("=" * 50)


# ─────────────────────────────────────────────
# STEP 2: APPLY 6 FILTERS
# ─────────────────────────────────────────────

# --- Filter 1: GRAYSCALE ---
# How it works: Mix R, G, B into one value
#   gray = 0.299*R + 0.587*G + 0.114*B
#   (Human eyes are most sensitive to green!)
img_gray = img.convert("L")               # "L" means Luminance (grayscale)
print("\nFilter 1: Grayscale applied")
print("   Formula: gray = 0.299*R + 0.587*G + 0.114*B")


# --- Filter 2: BLUR (Gaussian Blur) ---
# How it works: Each pixel becomes the AVERAGE of its neighbors
#   Imagine smearing paint with your finger → edges become soft
img_blur = img.filter(ImageFilter.GaussianBlur(radius=4))
print("\nFilter 2: Blur applied")
print("   Each pixel = average of surrounding pixels (radius=4)")


# --- Filter 3: EDGE DETECTION ---
# How it works: Find where colors CHANGE sharply
#   Flat areas → black (no change)
#   Boundaries → white (big change)
img_edges = img.filter(ImageFilter.FIND_EDGES)
print("\nFilter 3: Edge Detection applied")
print("   Bright = boundary found, Dark = flat area")


# --- Filter 4: BRIGHTNESS BOOST ---
# How it works: Multiply every pixel value by a factor
#   factor > 1 → brighter (multiply up)
#   factor < 1 → darker (multiply down)
enhancer = ImageEnhance.Brightness(img)
img_bright = enhancer.enhance(2.0)        # 2.0 = twice as bright
print("\nFilter 4: Brightness x2 applied")
print("   Every R, G, B value multiplied by 2.0")


# --- Filter 5: INVERT (Negative) ---
# How it works: new_pixel = 255 - old_pixel
#   White (255) → Black (0)
#   Black (0)   → White (255)
#   Like a photo negative!
img_array_inv = 255 - img_array           # NumPy makes this super easy!
img_invert = Image.fromarray(img_array_inv.astype(np.uint8))
print("\nFilter 5: Invert applied")
print("   Formula: new_pixel = 255 - old_pixel")


# --- Filter 6: SHARPEN ---
# How it works: Enhance edges and fine details
#   Opposite of blur — makes boundaries crisper
img_sharp = img.filter(ImageFilter.SHARPEN)
img_sharp = img.filter(ImageFilter.SHARPEN)  # Apply twice for stronger effect
print("\nFilter 6: Sharpen applied")
print("   Enhances edges and fine details")


# ─────────────────────────────────────────────
# STEP 3: SHOW ALL RESULTS IN ONE FIGURE
# ─────────────────────────────────────────────

fig, axes = plt.subplots(2, 4, figsize=(16, 9))
fig.patch.set_facecolor('#1a1a2e')
fig.suptitle("🎨 Image Processing — 6 Filters Explained", 
             fontsize=20, fontweight='bold', color='white', y=0.98)

images = [
    (img,         "1. ORIGINAL",         "Your starting image"),
    (img_gray,    "2. GRAYSCALE",         "gray = 0.299R + 0.587G + 0.114B"),
    (img_blur,    "3. GAUSSIAN BLUR",     "Each pixel = avg of neighbors"),
    (img_edges,   "4. EDGE DETECTION",    "Finds sharp color changes"),
    (img_bright,  "5. BRIGHTNESS x2",    "Every pixel value × 2.0"),
    (img_invert,  "6. INVERT / NEGATIVE", "new = 255 − old"),
    (img_sharp,   "7. SHARPEN",           "Enhances edges & detail"),
]

# Flatten axes
ax_flat = axes.flatten()

for i, (image, title, subtitle) in enumerate(images):
    ax = ax_flat[i]
    cmap = 'gray' if image.mode == 'L' else None
    ax.imshow(image, cmap=cmap)
    ax.set_title(title, color='white', fontweight='bold', fontsize=11, pad=6)
    ax.set_xlabel(subtitle, color='#aaaacc', fontsize=8)
    ax.set_xticks([])
    ax.set_yticks([])
    for spine in ax.spines.values():
        spine.set_edgecolor('#444466')
    ax.set_facecolor('#0d0d1a')

# Hide the last unused subplot
ax_flat[7].axis('off')

# Add a legend explaining pixels
legend_text = (
    "🔑  KEY CONCEPT\n\n"
    "Every image = grid of pixels\n"
    "Each pixel = [R, G, B]\n"
    "Each value  = 0 to 255\n\n"
    "  0  = no color (dark)\n"
    "255 = full color (bright)\n\n"
    "Filters just do\n"
    "MATH on these numbers!"
)
ax_flat[7].text(0.5, 0.5, legend_text,
    ha='center', va='center',
    fontsize=11, color='white',
    fontfamily='monospace',
    bbox=dict(boxstyle='round,pad=1', facecolor='#0d0d2a', edgecolor='#5555aa', linewidth=2),
    transform=ax_flat[7].transAxes)
ax_flat[7].set_facecolor('#1a1a2e')

plt.tight_layout()
plt.savefig("/Users/mayurimamdi/Downloads/image_processing_results.png", 
            dpi=130, bbox_inches='tight', facecolor='#1a1a2e')
print("\n" + "=" * 50)
print("DONE! Results saved to: image_processing_results.png")
print("=" * 50)


# ─────────────────────────────────────────────
# STEP 4: SHOW SOME PIXEL VALUES (Educational!)
# ─────────────────────────────────────────────
print("\nPIXEL VALUES EXAMPLE (top-left 3x3 area):")
print("   Original image (R, G, B):")
for row in img_array[:3, :3]:
    print("  ", [list(p) for p in row])

gray_array = np.array(img_gray)
print("\n   After Grayscale (single value):")
for row in gray_array[:3, :3]:
    print("  ", list(row))

inv_array = np.array(img_invert)
print("\n   After Invert (255 - original):")
for row in inv_array[:3, :3]:
    print("  ", [list(p) for p in row])

print("\n💡 Notice: Invert + Original = 255 always! (235+20=255) ✓")
```
